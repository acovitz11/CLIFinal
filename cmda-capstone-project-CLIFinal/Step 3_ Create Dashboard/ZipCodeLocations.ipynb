```python
import pandas as pd
import pgeocode

dataset = pd.read_csv("C:/Users/ashco/Downloads/FinalParishData.csv")
dataset.head()
```

```python
from pyzipcode import ZipCodeDatabase

# Initialize the ZipCodeDatabase
z = ZipCodeDatabase()

def get_state(zip_code):
    if pd.isna(zip_code) or not str(zip_code).isdigit():  # Check if the ZIP code is NaN or non-numeric
        return None
    try:
        zip_info = z[zip_code]  # Try to get the state from the database
        return zip_info.state
    except (KeyError, IndexError):  # Handle cases where ZIP code is not found
        return None

dataset['State'] = dataset['Zip'].apply(get_state)
dataset
```
